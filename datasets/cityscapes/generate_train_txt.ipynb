```python
import os
```

```python
path = os.getcwd()
```

```python
ls = os.listdir()
```

```python
ls.remove('generate_train_txt.ipynb')
ls.remove('.ipynb_checkpoints')
# ls.remove('train.txt')
```

```python
# ls
```

```python
path_dic = {}
path_dic_label = {}
train_dic = {}

for i in range(len(ls)):
    path_dic[ls[i]] = os.path.join(path,ls[i])
    path_dic_label[ls[i]] = os.path.join(path_dic[ls[i]],'Labels/')
    path_dic[ls[i]] = os.path.join(path_dic[ls[i]],'Images/')
for i in ls:
    train_dic[i] = []
for i in range(len(ls)):
    tmp1 = []
    for j,k in zip(os.listdir(path_dic[ls[i]]),os.listdir(path_dic_label[ls[i]])):
        tmp1.append(os.path.join(path_dic[ls[i]],j))
        tmp1.append(os.path.join(path_dic_label[ls[i]],k))
    train_dic[ls[i]] = tmp1
```

```python
train = (list(train_dic.values()))
f = open('train.txt', 'w')
x = 0
for i in train:
    for j in i:
        
        if x%2!=0:
            f.write(j)
            f.write('\n')
        else:
            f.write(j)
        x+=1
f.close()
```
